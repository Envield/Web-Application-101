```python
import pandas as pd
import numpy as np
import re
```

```python
comments_df = pd.read_csv('./dataset/comments_sentiment.csv')
comments_df.head(10)
```

```python
comments_df = comments_df.dropna(subset=['Comments'])
comments_df['Comments'] = comments_df['Comments'].str.replace(r'\byg\b', '', regex=True)
comments_df['Comments'] = comments_df['Comments'].str.replace(r'<a href="[^"]*">[^<]*</a>', '', regex=True)
comments_df.head(10)
```

```python
comments_df.to_csv('./dataset/cleaned_comments.csv', index=False)
```
